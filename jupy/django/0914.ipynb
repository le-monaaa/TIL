## Migration

Model 클래스의 변경사항(필드생성, 수정, 삭제 등)을 DB에 최종 반영하는 방법

model class(설계도 초안)-makemigrations->migration파일(최종 설계도)-migrate->db.sqlite3(DB)

이미 생성된 테이블에 필드를 추가해야 한다면?

model class 변경 -> makemigrations -> migrate

### Model field

DB 테이블의 필드(열)을 정의하며, 해당 필드에 저장되는 데이터 타입과 제약조건을 정의

CharField()

길이의 제한이 있는 문자열을 넣을 때 사용. max_length가 필수인자.

TextField()

제약조건X. 글자 수가 많을 때 사용

DateTimeField()

날짜와 시간을 넣을 때 사용. 
auto_now : 데이터가 저장될 때마다 자동으로 현재 날짜 시간을 저장(수정일)
auto_now_add : 데이터가 처음 생성될 때마다 자동으로 현재 날짜시간을 저장(생성일)



### Admin site

Automatic admin interface

Django는 추가 설치 및 설정 없이 자동으로 관리자 인터페이스를 제공.

-> 데이터 확인 및 테스트 등을 진행하는데 유용히 사용 가능.

python manage.py createsuperuser

admin에 모델 클래스 등록
from import 해주고
admin.site.register(클래스)

데이터베이스 초기화

migration파일 삭제(번호붙은 파일만)

db.sqlite3 파일 삭제. (_\_init__.py 삭제 x, migrations폴더 삭제 x) 

python manage.py showmigrations -> migration이 되었는지 확인 [X] : migrate o

python manage.py sqlmigrate articles 0001 ->

-> 해당 migrations파일이 sql 언어로 어떻게 번역되어 DB로 전달되는지 확인하는 명령어.


SQLite : 데이터베이스 관리 시스템 중 하나이며 django의 기본 데이터베이스로 사용됨(파일로 존재하며 가볍고 호환성이 좋음)
