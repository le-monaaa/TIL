```python
<!DOCTYPE html>
<html lang="en">

<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Document</title>
</head>

<body>
  <div id="app">
    <h1>{{greeting}} </h1>
    <button @click="count++">{{count}} </button>
    <p>{{count}} </p>
    <h3>{{count}} </h3>
    <!-- vanilla js     Vue          -->
  </div>

  <script src="https://unpkg.com/vue@3/dist/vue.global.js"></script>
  <script>
    // JS  - "  "
    const { createApp, ref } = Vue
    // vue application instance  
    const app = createApp({
      //      method(setup  ) .
      // " " 
      setup() {
        // ref ->   vue   
        const count = ref(0)

        const greeting = ref("hello")

        //   (key:value )
        // setup method return   template   
        return {
          // " " key  value       .
          count,
          greeting
        }
      }
    })
    //     mount  ,
    app.mount("#app")

  </script>
</body>

</html>
```

```python
<!DOCTYPE html>
<html lang="en">

<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Document</title>
</head>
  <div id="app">
    <!-- v-on:(click)  =""    -->
    <button v-on:click="increment"> {{count}} </button>
  </div>
<body>
  <script src="https://unpkg.com/vue@3/dist/vue.global.js"></script>
  <script>
    const {createApp, ref} = Vue

    const app = createApp({
      setup() {
        const count = ref(0)
        // count  
        const increment = function () {
          // count++
          // ref     (const  ) 
          // ref       ( )
          count.value++
        }
        return{
          count,
          increment
        }
      }
    })

    app.mount("#app")

  </script>
</body>

</html>
```
