1부터 M까지 자연수 중에서 중복없이 M개를 고른 수열. 사전 순으로 증가하는 순서로 출력.

구분: 백트래킹

**백트래킹**
0814, 0918 정리 참고.
 여러 옵션 중 하나를 선택, 선택이 이루어지며 선택지들의 집합이 생성되고 이러한 생성을 반복하며 최종 상태에 도달.
 재귀적으로 문제를 하나씩 풀어나가다 현 노드에 유망성이 없다 판단될 경우 제외하고 다음 노드 탐색으로 넘어감.

**백트래킹과 DFS의 차이**

백트래킹은 가지치기를 함. dfs는 모든 경로를 추적하지만 백트래킹은 불필요한 경로를 조기에 차단함. 백트래킹 적용시 일반적으로 경우의 수가 줄어드나, 최악의 경우 지수함수시간을 요하므로 처리 불가.
