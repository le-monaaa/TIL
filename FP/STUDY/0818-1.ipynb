# BFS Breadth First Search 

- 너비우선탐색.

    탐색 시작점의 인접한 정점들을 먼저 모두 차례로 방문한 후에 방문했던 정점을 시작점으로 해 다시 인접한 정점을 차례로 방문하는 방식

    인접한 정점들에 대해 탐색을 한 후, 차례로 다시 너비우선탐색을 진행하므로, 선입선출 형태의 큐를 사용

    탐색 시작 노드를 큐에 삽입하고 방문처리
    
    큐에서 노드를 꺼낸 후 해당 노드의 인접 노드 중 방문하지 않은 노드를 모두 큐에 삽입, 방문 처리

    위의 과정을 수행할 수 없을 때까지 반복

    사용 : 두 노드 사이의 최단 경로 혹은 임의의 경로를 찾고 싶을 때 이용

    DFS보다 조금 더 복잡

    직관적이지 않은 면- BFS는 시작 노드에서 시작해 거리에 따라 단계별로 탐색

    BFS는 재귀적으로 동작하지 않음(짱조앟)

    Prim, Dijkstra 알고리즘과 유사

**BFS 예제2**







```python
# BFS 샘플 코드

def bfs(arr, start, visited):
    q = []
    visited[start] = True
    while q:
        v = q.pop()
        for i in arr[v]:
            if not visited[i]:
                q.append(i)
                visited[i] = True
            
```

**\+ DFS의 장단점**

장점 : 현재 경로상의 노드들만 기억하면 되므로 저장 공간의 수요가 비교적 작다

    목표 노드가 깊은 단계에 있을 경우 비교적 해를 빨리 구할 수 있다

단점 : 해가 없는 경로에 깊이 빠질 가능성

    얻어진 해가 최단 경로가 된다는 보장이 없다'


dfs는 bfs보다 좀 더 간단하게 구현 가능. 검색 속도는 dfs가 bfs보다 느림.

bfs는 최단 거리 문제에 효율, dfs는 이동한 정점의 값을 가지고 계속 연산하는 경우(재귀호출)에 효율

```python
def bfs(graph, start):
    visited = [] # 방문할 노드를 담을 배열
    q = [] # 정점과 인접한 방문 예정 노드를 담을 배열

    q.append(start) # 시작 노드 담고 시작

    while q: # 더 이상 방문할 노드가 없을 때까지
        node = q.pop(0) # 방문할 노드를 앞에서부터 하나씩 꺼냄

        if node not in visited: # 방문한 노드가 아니라면
            visited.append(node) # visited 배열에 추가
            q.extend(graph[node]) # node의 자식 노드도 추가
    
    print("bfs - ", visited)
    return visited

gr = {
    "A" : ["B"],
    "B" : ["A", "C", "H"],
    "C" : ["B", "D"],
    "D" : ["C","E","G"],
    "E" : ["D", "F"],
    "F" : ["E"],
    "G" : ["D"],
    "H" : ["B", "I", "J", "M"],
    "I" : ["H"],
    "J" : ["H", "K"],
    "K" : ["J", "L"],
    "L" : ["K"],
    "M" : ["H"]
}

bfs(gr, "G")
```
