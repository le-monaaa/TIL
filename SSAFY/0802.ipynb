델타를 가지고 문제해결이 익숙해질떄까지,

```python
N = int(input())
arr = [list(map(int, input().split())) for _ in range(N)]
total1 = 0
total2 = 0
for i in range(N):
    total1 += arr[i][i]
    total2 += arr[i][N-i]
```
