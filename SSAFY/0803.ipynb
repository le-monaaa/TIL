```python
def selectionSort(a, N): # a:  , N: a  
    for i in range(N-1): #    
        minIdx = i
        for j in range(i+1, N):
            if a[minIdx] > a[j]: #  minIdx  .
                minIdx = j
        a[i], a[minIdx] = a[minIdx], a[i] # 
```
