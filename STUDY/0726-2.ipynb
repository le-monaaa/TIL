### 패키지(package) = library

- 특정 기능과 관련된 여러 모듈을 한 그룹으로 묶은 것
- 패키지 안 패키지 가능
- import module from package (import package)

### 모듈(module)
- 여러 기능들이 뭉쳐진 하나의 .py파일
- 함수, 클래스, 변수 등 포함
- import module

### 함수(function)
- 하나의 기능을 가진 코드의 집합
- 함수를 사용한다 = 함수를 호출한다
- 함수 사용 형태 : 함수()
  필요한 경우 ()안 매개변수 또는 옵션 들어감
- 종류: 내장함수, 외장함수(import)

### 클래스(class)
- 파이썬은 객체 지향 언어로서 클래스를 기반으로 객체를 만들어 사용
- .py 내에서 "class class_name():"으로 선언

패키지>모듈>클래스>함수
