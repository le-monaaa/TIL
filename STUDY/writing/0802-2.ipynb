### 시간 복잡도

기본적인 연산을 수행하는 데에 어떤 고정된 시간이 걸릴 때,

알고리즘에 의해서 수행되는 기본 연산의 갯수를 세어 예측할 수 있다.

그러므로 걸리는 시간의 총량과 알고리즘에 의해 수행되는 기본적인 연산의 갯수는 최대 상수 인자만큼 다르다.

--> 연산의 갯수를 세어 얼마만큼의 연산이 수행되는가를 통해 로직의 효율을 분석하는 데 사용됨

#### 시간 복잡도 표기법

- Big-O 표기법 : 최악의 실행 시간. 최소한 보장되는 성능 표기

- Big-Ω 표기법 : 최상의 실행 시간.

- Big-θ 표기법 : 평균 실행 시간.

#### Big-O 표기법

- 특징 :

시간 복잡도에 미미한 영향을 주는 것들은 배제

1. 상수항 무시: O(N+5)일 경우, 상수 생략, O(N)으로 표시

2. 계수 무시: O(3N)일 경우, 계수 생략, O(N)으로 표시

3. 최고차항만 표기: O(3N^3+2N^2+N+5)일 경우 O(N^3)으로 표기

##### Big-O 표기법 종류



1. O(1)

- 입력값(N)이 증가해도 실행시간은 동일.

- index로 접근해 바로 처리할 수 있는 연산 과정의 시간 복잡도->기본연산수

- ex) stack의 push, pop

2. O(log N)

- 연산이 한 번 실행될 때 마다 데이터의 크기가 절반 감소.(log 지수 2)

- ex) binary search 알고리즘, tree 형태 자료구조 탐색

3. O(N)

- 입력값(N)이 증가함에 따라 실행시간도 선형적으로 증가하는 알고리즘

- ex) 1중 for문

4. O(N log N)

- O(N)의 알고리즘과 O(log N)의 알고리즘이 중첩된 형태

5. O(N^2)

- O(N)의 알고리즘과 O(log N)의 알고리즘이 중첩된 형태

- ex) 2중 for문, 삽입/버블/선택 정렬

6. O(2^N)

- 재귀적으로 수행하는 알고리즘

ex) 피보나치 수열
