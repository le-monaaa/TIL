- 바이트 순서에 대해 표준화하지 못했다(파일 인식시 UCS-2, UCS-4인지 먼저 인식하고 각 경우에 대해 구분해서 구현해야하는 불편)

- 유니코드의 외부 인코딩이 필요하게 됨 -->

**BOM(Byte Order Mark)**
 -big-edian, little-endian-
 문서 맨 앞에 눈에 보이지 않는 byte를 넣고 이를 해석해서 정확히 어떤 인코딩 방식이 사용되었는지 알아내는 방법. 유니코드가 big-ednian인지 big-endian인지 또는 UTF-8인지 알 수 있도록 유니코드 파일 시작 첫부분에 보이지 않게 2~3바이트의 문자열 추가. --> BOM
 텍스트 에디터 화면에서는 보이지 않고 헥사 에디터로 확인 가능.

 UTF-8: EF BB BF
 UTF-16 big-endian : FE FF
 UTF-16 little-endian : FF FE 
 이런식으로~~

문제점 : UTF-8에는 BOM이 없는 것이 보통인데(UTF-8은 BOM이 고정이라 인코딩 방식을 자동으로 알 수 있음) 일부 윈도우즈 프로그램(메모장 등)은 UTF-8 파일 생성시 자동으로 BOM을 집어넣어버림. 윈도우즈 환경에서는 문제가 없는 경우가 많지만 리눅스나 유닉스 환경에서는 문제가 됨. DB에서 BOM이 추가된 데이터와 그렇지 않은 데이터를 비교할 경우 눈으로는 동일하지만 사실 같지 않은 데이터인 경우 많다 ---> 이게 문제

해결: 메모장같은 프로그램보다 BOM설정이 가능한 프로그램을 사용. 하지만 모든 파일에 대해 해당 프로그램을 사용할 수 없기 때문에 BOM파일을 받은 후 코드로 해당 파일을 UTF-8로 인코딩 처리하는 방법도 있음.

int()와 같은 atoi()함수 만들기



연산 과정을 이해해 볼 필요가 있다


```python
# str() 함수를 사용하지 않고 itoa()를 구현해보자
# 양의 정수를 입력받아 문자열로 반환하는 함수
# 입력값: 변환할 정수 값, 반환된 문자열을 저장할 문자배열
# 반환값: 없음
# ord(), chr() 사용

def itoa(a):
    s = ''
    while a>0:
        s += chr(ord('0') + a % 10) # 1의자리의 ascii코드 값을 받음
        a //= 10 # 이미 구한 1의자리 를 제외한 나머지에 대해~~
    # return s # 뒤집어서 나옴
    return s[::-1]
a = 123
print(itoa(a)) 
```
