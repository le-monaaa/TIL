#### 문자열

##### 문자열 조회/탐색 메서드

```python
st = "Cozy"
st1 = "12312412"

# s.find(x) : x의 첫 번째 위치를 반환, 없으면 -1 반환
st.find("z") #2

# s.index(x) : x의 첫 번째 위치를 반환, 없으면 오류 발생
st.index("y") #3
# st.index("i") #ValueError

# s.isaplpha() : 알파벳 문자 여부 *단순 알파벳이 아닌 유니코드 상의 Letter(한국어포함)
st.isalpha() # True
st1.isalpha() # False

# s.isupper() : "전체"문자열의 대문자 여부 
st.isupper() #False
st1.isupper() #False

# s.islower() : "전체"문자열의 소문자 여부
st.islower() # False
st1.islower() # False

# s.istitle() : 문자열 안 각 단어의 첫 글자만 대문자인지 체크. 
# 특수문자 뒤의 단어도 대문자로 시작해야함, 숫자 및 특수문자 무시.
st2 = "I Have 3cookies"
st2.istitle() # False. 3뒤의 뛰어쓰기 되지 않은 글자도 대문자 체크.
st3 = "Naver.com"
st3.istitle() #False. 특수문자 뒤 단어도 대문자로 시작해야함
```

##### 문자열 조작 메서드
BNF표기법 알아보기
파이썬은 EBNF 확장된 BNF

```python
# s.replace(old, new[, count]) : 바꿀 대상 글자를 새로운 글자로 바꿔서 변환
"Hello, world!".replace("world", "python") #Hello, python!

# s.strip([chars]) : 문자열의 시작과 끝에 있는 공백 혹은 지정한 문자 제거
"Hello, world!     ".strip() # Hello, world!
"Hello, world!".strip("r") # 중간의 문자는 사라지지않는다.
# lstrip() 왼쪽 끝/ rstrip() 오른쪽 끝

# s.split() : 지정한 문자를 구분자로 문자열을 분리해 문자열 리스트로 반환
"Hello, world!".split() # 기본 설정은 공백으로.
"Hello, world!".split("w") # ['Hello, ', 'orld!']

# "seperator".join([iterable]) : iterable 요소들을 원래의 문자열을 구분자로 하나의 문자열로 연결
words = ["Hello", "world!"]
text = "-".join(words)
print(text) # Hello-world!

"Hello, world!".upper() #HELLO, WORLD! s.upper() : 문자열을 모두 대문자로 바꿔줌
"hello, world!".capitalize() # s.capitalize() : 첫글자를 대문자로 바꿔줌
"hEllo, WoRLd!".swapcase() # s.swapcase() : 소문자 대문자를 변환해줌
"helLo, woRld!".title() # 문자열 내 띄어쓰기 기준으로 각 단어의 첫글자는 대문자로, 나머지는 소문자로 변환

# 메서드는 이어서 사용가능하다.
text2 = "heLLo, woRld!"
new_text = text2.swapcase().replace("l", "z")
print(new_text) #HEzzO, WOrLD!
```

#### 리스트 관련 메서드
#### 리스트 값 추가 및 삭제 메서드

```python
# .append(x) : 리스트 마지막에 항목 x를 추가
# .extend(x) : iterable x의 모든 항목들을 리스트 끝에 추가. +=와 같음
# append 와 extend의 차이
# append로 리스트 m를 추가하면 리스트 m 자체가 원 리스트의 인덱스 한 부분으로 들어감,
# extend는 리스트 m의 요소를 원 리스트에 추가함.

# .insert(i, x) : 리스트 인덱스 i에 항목 x를 삽입

# .remove(x) : 리스트 가장 왼쪽에 있는 항목 (첫 번째) x를 제거. 없을경우 ValueError

# .pop() : 리스트 가장 오른쪽에 있는 항목(마지막)을 반환 후 제거
# 삭제되는 항목을 변수에 저장할 수 있다.
alist = [1, 2, 3, 4, 5]
item1 = alist.pop()
print(item1) # 5

# .pop(i) : 리스트 인덱스 i에 있는 항목을 반환 후 제거

# .clear() : 리스트의 모든 항목 삭제
```

#### 리스트 탐색 및 정렬 메서드


```python
# .index(x, start, end) : 리스트에 있는 항목 중 가장 왼쪽에 있는 항목 x의 인덱스
# .reverse() : 리스트를 거꾸로 변경. 정렬 X

# .sort() : 리스트 정렬. 매개변수 사용 가능
# .sorted() : 원 리스트 변경x, 정렬된 새로운 리스트 반환해줌. 딕셔너리 객체도 받을 수 있음
# 기본:오름차순/ sorted(reverser=True) : 내림차순.
# sort랑 sorted 헷갈린다니까 확실히!! 

data = [["가", 1], ["다", 3], ["나", 2]]
print(sorted(data, key = lambda x: x[0])) # 0번째 인수 기준 정렬
print(sorted(data, key = lambda x : x[1])) # 1번째 인수 기준 정렬
print(sorted(data, reverse = True, key = lambda x : x[1])) #1번째 인수 기준으로 내림차순 정렬

# 불변형인 tuple 을 정렬하고싶다면 sorted(튜플) 을 이용할 수 있다. 문자열도 정렬이 됨. 
# 딕셔너리는 키 기준. 값을 기준으로 정렬하고싶다면 ? reverse

dic = { "a" : 3, "b" : 23, "c" : 10}

def sort_key(item):
    #정렬 기준을 return
    print(item)
    return item[1] # 0일때 key 기준, 1일때 value 기준 정렬

print(sorted(dic.items(), key = sort_key))



# .count() : 리스트에서 항목 x의 개수를 반환

# .isdecimal() : 문자열이 모두 순자 문자(0~9)로만 이루어져 있어야 True
# .isdigit() : 유니코드 숫자도 인식(원문자숫자도 숫자로 인식)
# .isnumeric() : 몇 가지 추가적인 유니코드까지 인식. (분수, 지수, 루트기호도 숫자로 인식)
```

startswith() : 접두사로 문장 찾기. 정의된 문자열이 지정된 접두사로 시작하면 True 반환.
str.startswith(접두사[, start [,end]])
접미사는 찾고자 하는 접미사들의 튜플도 가능.
end(선택) 사용하면 해당 위치에서 비교 중단.

```python
numbers = [1, 2, 3, 4, 5]
alist = numbers

# 전체 슬라이싱

list2 = numbers[:]

numbers[0] = 100

print(alist) # [100, 2, 3, 4, 5] #복제 리스트의 값도 영향을 받음
print(list2) # [1, 2, 3, 4, 5] # 슬라이싱으로 가져온 리스트는 변하지 않음
#저장된 주소값이 다르다., 슬라이싱은 복사본을 생성. 참조하는게아님

#파이썬튜터로 설명
```
